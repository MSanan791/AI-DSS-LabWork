```python
class Node:
    def __init__(self, value):
        self.value = value
        self.left = None
        self.right = None

class BST:
    def __init__(self, root = None):
        self.root = root
    def insert(self, value):
        if self.root is None:
            self.root = Node(value)
        else:
            self.insert_recursive(self.root, value)
            return
    def insert_recursive(self, root, value):
        if value < root.value:
            if root.left is None:
                root.left = Node(value)
            else:
                self.insert_recursive(root.left, value)
        elif value > root.value:
            if root.right is None:
                root.right = Node(value)
            else:
                self.insert_recursive(root.right, value)
        return
    def search(self,value):
        return self.search_recursive(self.root, value)

    def search_recursive(self, root, value):
        if root is None:
            return False
        if value == root.value:
            return True
        if value < root.value:
            return self.search_recursive(root.left, value)
        else:
            return self.search_recursive(root.right, value)


bst = BST()
bst.insert(10)
bst.insert(5)
bst.insert(15)
bst.insert(2)
bst.insert(7)
bst.insert(12)

# After insertions, the tree looks like this:
#       10
#      /  \
#     5    15
#    / \   /
#   2   7 12

print(f"Does the tree contain 15? {bst.search(15)}")  # Expected: True
print(f"Does the tree contain 7? {bst.search(7)}")    # Expected: True
print(f"Does the tree contain 99? {bst.search(99)}")  # Expected: False

```
