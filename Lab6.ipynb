```python
class Node:
    def __init__(self, value = [None,None]):
        self.value = value
        self.node_left = None
        self.node_right = None

class Tree:
    """
    The basic data type is Node and Node is storing values in tuples
    Node = (root_id,value)
    """
    def __init__(self, root = None):
        self.root = root
    def insert(self, node):
        self.root = node
    def search_node(self, root_id ='string', node = None):
        if node is None:
            return None
        if node.value[0] == root_id :
            return node
        cur_node = self.search_node(root_id, node.node_left)
        if cur_node is None:
            cur_node = self.search_node(root_id,node.node_right)
        return cur_node
    def search(self,root_id):
        return self.search_node(root_id,self.root)

    def insert_below_node(self,root_id, value_id, value):
        node = self.search_node(root_id,self.root)
        if node is None:
            print(f'Parent Node with ID {root_id} was not found, cannot insert value here  ')
            return
        if node.node_left is None:
            node.node_left = Node([value_id,value])
            print(f"NODE ADDED SUCCESSFULLY Node-> {node.node_left}")
        elif node.node_right is None: node.node_right = Node([value_id,value])
        else:
            print('Value already exists at the given position')

        return

    def print_tree(self,node):
         if node is None:
            return None
         cur_node = self.print_tree(node.node_left)
         if cur_node is None:
             cur_node = self.print_tree(node.node_right)
         print(node.value)
         return

    def print_self(self):
        self.print_tree(self.root)
        return


```

Creating the MinMax function

```python
def minimax(node,depth):

    if node.node_left is None and node.node_right is None:
        return node.value[1]
    if depth % 2 == 0:
        best_value = float('-inf') # Initialize with negative infinity

        if node.node_left is not None:
            left_value = minimax(node.node_left, depth +1)
            best_value = max(best_value,left_value)

        if node.node_right is not None:
            right_value = minimax(node.node_right, depth +1)
            best_value = max(best_value,right_value)

        node.value[1] = best_value
        return best_value
    else:
        best_value = float('inf')
        if node.node_left is not None:
            left_value = minimax(node.node_left, depth+1)
            best_value = min(best_value, left_value)
        if node.node_right is not None:
            right_value = minimax(node.node_right, depth+1)
            best_value = min(best_value,right_value)

        node.value[1] = best_value
        return best_value
```

```python
final_score = minimax(tree1.root, 0)
print(f"The optimal score for the root node is: {final_score}")
print(f"The root node's updated value is: {tree1.root.value}")

```

```python
def minmax_ab(node, depth, alpha,beta):
    if node.node_left is None and node.node_right is None :
        return node.value[1]
    if depth % 2 == 0: # for even values
        best_value = float('-inf')
        if node.node_left is not None:
            value = minmax_ab(node.node_left,depth+1, alpha,beta)
            best_value = max(best_value,value)
            alpha = max(alpha,best_value)
            if alpha>= beta:
                return best_value
        if node.node_right is not None:
            value = minmax_ab(node.node_right, depth+1,alpha,beta)
            best_value = max(best_value,value)
            alpha = max(alpha,best_value)
            if alpha>=beta:
                return  best_value
        node.value[1] = best_value
        return best_value
    else:
        best_value = float('inf')
        if node.node_left is not None:
            value = minmax_ab(node.node_left, depth+1, alpha, beta)
            best_value = min(best_value,value)
            beta = min(best_value,beta)
            if alpha >= beta:
                return best_value

        if node.node_right is not None:
            value = minmax_ab(node.node_right, depth+1, alpha, beta)
            best_value = min(value, best_value)
            beta = min(best_value, beta)
            if alpha >= beta:
                return best_value
        node.value[1] = best_value
        return best_value
```

```python
initial_alpha = float('-inf')
initial_beta = float('inf')

final_score_ab = minmax_ab(tree1.root, 0, initial_alpha, initial_beta)
print(f"The optimal score with Alpha-Beta Pruning is: {final_score_ab}")
print(f"The root node's updated value is: {tree1.root.value}")
```
