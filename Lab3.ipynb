### 1.	Implement the Breadth-First Search (BFS) traversal using a FIFO (queue) data structure for the given graph, starting the traversal from node ‘C’.

```python
alphabet_graph = {
    'E': ['A','B'],
    'A': ['E','B','D'],
    'B': ['E','A','D'],
    'D': ['A','B','C'],
    'C': ['D']
}

def bfs(graph, start):
    visited = set()  # Set to keep track of visited nodes.
    queue = [start]  # Initialize a queue with the start node.

    while queue:
        vertex = queue.pop(0)  # Dequeue a vertex from the queue
        if vertex not in visited:
            print(vertex, end=" ")  # Process the vertex (print it)
            visited.add(vertex)  # Mark the vertex as visited
            queue.extend([node for node in graph[vertex] if node not in visited])  # Enqueue unvisited neighbors
    print()  # For a new line after traversal
print("BFS traversal starting from node 'C':")
bfs(alphabet_graph, 'C')
```

## 3.	Implement a LIFO (stack) data structure in Python and use it to perform the Depth-First Search (DFS) algorithm both recursively and iteratively. Apply the implemented DFS algorithm to traverse Graph 1 (starting from node ‘6’) and Graph 2 (starting from node ‘E’).

```python
def dfs_recursive(graph, vertex, visited=None):
    if visited is None:
        visited = set()
    visited.add(vertex)
    print(vertex, end=" ")  # Process the vertex (print it)
    for neighbor in graph[vertex]:
        if neighbor not in visited:
            dfs_recursive(graph, neighbor, visited)
    return visited

un_graph = {
    6: [4],
    4: [3,5],
    5: [4,2,1],
    3: [3,5],
    2: [5,1],
    1: [2,5]
}

un_graph2 = {
    'E': ['A','B'],
    'A': ['E','B','D'],
    'B': ['E','A','D'],
    'D': ['A','B','C'],
    'C': ['D']
}

print("\nDFS Recursive traversal starting from node '6':")
dfs_recursive(un_graph, 6)
print()  # For a new line after traversal

print("\nDFS Recursive traversal starting from node 'E':")
dfs_recursive(un_graph2, 'E')
print()  # For a new line after traversal
```

## 4.	Perform Depth-First Search (DFS) traversals on Tree 1 and Tree 2 using Preorder, Inorder, and Postorder methods. Start the traversal from node ‘1’ for Tree 1 and from node ‘50’ for Tree 2.


```python
tree_1 = {
    1: [2,3],
    2: [4,5],
    4: [6],
    5: [7,8],
    3:[],
    6:[],
    7:[],
    8:[]
}

tree_2 = {
    50:[17,76],
    17:[9,23],
    9:[14],
    12:[],
    23:[19],
    76:[54],
    54:[72],
    72:[67],
    67:[],
    19:[],
    14:[12]
}

def dfs_preorder(graph, vertex, visited=None):
    if visited is None:
        visited = set()
    visited.add(vertex)
    print(vertex, end=" ")  # Process the vertex (print it)
    for neighbor in graph[vertex]:
        if neighbor not in visited:
            dfs_preorder(graph, neighbor, visited)
    return visited
def dfs_inorder(graph, vertex, visited=None):
    if visited is None:
        visited = set()
    visited.add(vertex)
    neighbors = graph[vertex]
    if neighbors:
        dfs_inorder(graph, neighbors[0], visited)  # Visit left child
    print(vertex, end=" ")  # Process the vertex (print it)
    for neighbor in neighbors[1:]:
        if neighbor not in visited:
            dfs_inorder(graph, neighbor, visited)  # Visit right child
    return visited
def dfs_postorder(graph, vertex, visited=None):
    if visited is None:
        visited = set()
    visited.add(vertex)
    for neighbor in graph[vertex]:
        if neighbor not in visited:
            dfs_postorder(graph, neighbor, visited)
    print(vertex, end=" ")  # Process the vertex (print it)
    return visited
print("\nDFS Preorder traversal starting from node '1':")
dfs_preorder(tree_1, 1)
print()  # For a new line after traversal
print("\nDFS Inorder traversal starting from node '1':")
dfs_inorder(tree_1, 1)
print()  # For a new line after traversal
print("\nDFS Postorder traversal starting from node '1':")
dfs_postorder(tree_1, 1)
print()  # For a new line after traversal
print("\nDFS Preorder traversal starting from node '50':")
dfs_preorder(tree_2, 50)
print()  # For a new line after traversal
print("\nDFS Inorder traversal starting from node '50':")
dfs_inorder(tree_2, 50)
print()  # For a new line after traversal
print("\nDFS Postorder traversal starting from node '50':")
dfs_postorder(tree_2, 50)
print()  # For a new line after traversal
```
